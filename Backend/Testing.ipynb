```python
import yfinance as yf
import pandas as pd
```

```python
ticker = yf.Ticker("AAPL")
type(ticker.info                               ) # dict
type(ticker.calendar                           ) # dict
type(ticker.analyst_price_targets              ) # dict
type(ticker.quarterly_income_stmt              ) # df
type(ticker.history(period='1mo')              ) # df
type(ticker.option_chain(ticker.options[0]).calls) # df
```

```python
def is_valid_ticker(ticker: str) -> bool:
    try:
        t = yf.Ticker(ticker)
        return len(t.calendar) > 0
    except Exception:
        return False

print(is_valid_ticker("aapl"))
```
